```python
#Correcting reference databases which have multiple peaks in a single cell (and therefore can't immediately be turned into float)
import re

def correct_value_type(value):
    try:
        float(value)
    except:
        #try to resolve, check if there are multiple values with a delimiter, e.g. /, in one cell
        delimiters = "/", "-", " or ", "•"
        #allow for all possible delimiters
        regex_pattern = '|'.join(map(re.escape, delimiters))
        #split into separate values
        attempt_split = re.split(regex_pattern, value)
        #better if I could print the exact location in the dataframe in place
        print("Multiple reference values found for one cell: {}, using {}.".format(attempt_split, attempt_split[0]))
        if len(attempt_split)>1:
            #currently, just take first value
            value = attempt_split[0]
            #recursive, to double-check the listed value is a float
            correct_value_type(value)
        else:
            raise ValueError("Not able to interpret elements of {} as floats. \n Try adding any unusual characters to the 'delimiters' variable in the function correct_value_type().".format(attempt_split))
            
        
    return float(value)

ref_peaks = reference_raw.map(correct_value_type)
```

## Run script to identify species and peptide markers in one sample


```python
# Function with one sample (what was saved as df)
def transform(df):
    df_copy = df.copy()  # making a copy so I dont have to reload the dataframe everytime
    x = np.arange(500, 3500, 0.1)  # Define the x-axis range

    resampled = np.zeros_like(x)  # We initialise the 'resampled' array
    for p in df_copy['mass']:
        resampled[np.logical_and(p-x > -0.3, p-x < 1.3)] = 1  # This is the default range (error range), but we specify for which peptides later

    resampled_smooth = gaussian_filter1d(resampled, 100)  # smoothing
    plt.plot(x, resampled, label='sample', linewidth=1)
    plt.plot(x, resampled_smooth, label='sample', linewidth=1)

    n_spec = ref_peaks.shape[0]  
    corrs = np.zeros(n_spec)  

    for i in range(n_spec):
        y = np.zeros_like(x)
        pi = np.array(ref_peaks.iloc[i].to_numpy())
        pi = pi[~np.isnan(pi)]  # Remove NaN values from reference peaks

        for j, p in enumerate(pi):
            if ref_peaks.columns[j] in ['COL1a1_586___618', 'COL1a1_586___618_16', 'COL1_1_508_519', 'COL1a2_502___519', 'COL1a2_793___816']:
                # Custom thresholds for specific the specific peptide markers that deamidate
                y[np.logical_and(p-x > -0.3, p-x < 1.3)] = 1
            else:
                # Default thresholds for all other peptide markers
                y[np.logical_and(p-x > -0.3, p-x < 0.3)] = 1

        corrs[i] = np.corrcoef(y, resampled)[0, 1]  # We calculate the correlation

    ranks = pd.DataFrame({'Order': reference['Order'],
                          'Family': reference['Family'], 
                          'Species': reference['Species'], 
                          'ZooMS_taxon': reference['ZooMS_taxon'], 
                          'Correlation': corrs})

    idx = ranks.sort_values(by='Correlation', ascending=False).index[:6] - 1 
    # index [6] because the first 6 matches of species will be determined -- number depends on how many of the first n most likely species you want to look at

    all_d = pd.DataFrame(columns=ref_peaks.columns)
    for i in idx:
        y = np.zeros_like(x)
        pi = np.array(ref_peaks.iloc[i].to_numpy())
        pi = pi[~np.isnan(pi)]

        for j, p in enumerate(pi):
            if ref_peaks.columns[j] in ['COL1a1_586___618', 'COL1a1_586___618_16', 'COL1_1_508_519', 'COL1a2_502___519', 'COL1a2_793___816']:
                # Custom thresholds for peptide markers that contain Q or N
                y[np.logical_and(p-x > -0.3, p-x < 1.3)] = 1
            else:
                # Default thresholds for all other peptide markers
                y[np.logical_and(p-x > -0.3, p-x < 0.3)] = 1

        plt.figure()
        plt.plot(x, y * resampled)
        plt.title(reference['Species'][i+1])

        v, ix, iy = np.intersect1d((np.round(pi, decimals=1) * 10).astype(int), 
                                   (x[y * resampled == 1] * 10).astype(int), return_indices=True)

        d = pd.DataFrame({k: p for k, p in zip(ref_peaks.columns[ix], v / 10)}, 
                         index=pd.MultiIndex.from_tuples([(reference['Species'][i+1], 
                                                           reference['Family'][i+1], 
                                                           reference['Order'][i+1], 
                                                           reference['ZooMS_taxon'][i+1])],
                                                          names=['Species', 'Family', 'Order', 'ZooMS_taxon'])).reset_index()

        all_d = pd.concat([all_d, d], axis=0)

    ranks_df = ranks.sort_values(by='Correlation', ascending=False)

    return all_d
```
